```python
import pandas as pd

df = pd.read_csv("creditcard.csv")

print("Shape:", df.shape)
print("\nClass Distribution:\n", df['Class'].value_counts())
```

```python
df['Class'].value_counts(normalize=True) * 100
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("Class", axis=1)
y = df["Class"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train fraud ratio:", y_train.mean())
print("Test fraud ratio:", y_test.mean())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=500, solver="liblinear"))
])

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
results = []

results.append({
    "Model": "Logistic Regression (No Sampling)",
    "ROC-AUC": roc_auc_score(y_test, y_prob)
})
```

```python
print("Training started...")
pipeline.fit(X_train, y_train)
print("Training finished.")
```

```python
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:,1]

print("Predictions done")
```

```python
from sklearn.metrics import classification_report, roc_auc_score

report = classification_report(y_test, y_pred)
roc = roc_auc_score(y_test, y_prob)

print("Classification Report:\n")
print(report)

print("\nROC-AUC:", roc)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

from sklearn.metrics import roc_auc_score
print("Random Forest ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

print("Training started...")
rf.fit(X_train, y_train)
print("Training finished.")

y_prob_rf = rf.predict_proba(X_test)[:,1]
print("Random Forest ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
```

```python
rf_weighted = RandomForestClassifier(
    n_estimators=100,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

print("Training weighted RF...")
rf_weighted.fit(X_train, y_train)

y_prob_rf_w = rf_weighted.predict_proba(X_test)[:,1]

print("Weighted RF ROC-AUC:", roc_auc_score(y_test, y_prob_rf_w))
```

```python
from sklearn.metrics import classification_report

y_pred_rf_w = rf_weighted.predict(X_test)
print(classification_report(y_test, y_pred_rf_w))
```
